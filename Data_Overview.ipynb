### Displaying the Tags of Interest

```python
tags_split = df['interest_tags'].dropna().astype(str).str.split(', ')

tags = tags_split.explode().str.strip().value_counts()

print("--- Most Popular Interest Tags ---")
print(tags)
```

```python
tags_split = df['interest_tags'].dropna().astype(str).str.split(', ')
tags = tags_split.explode().str.strip().value_counts().reset_index()
tags.columns = ['interest_tags', 'interest_tags_count']

other_columns = [
    'gender', 'sexual_orientation', 'location_type', 'income_bracket', 
    'education_level', 'app_usage_time_label', 'swipe_right_label', 
    'swipe_time_of_day', 'match_outcome'
]

dfs_to_concat = [tags]

for col in other_columns:

    if col in df.columns:
    
        counts = df[col].value_counts().reset_index()
        counts.columns = [col, f'{col}_count']
        dfs_to_concat.append(counts)

summary_table = pd.concat(dfs_to_concat, axis=1)

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("--- Demographics & Usage Summary Table ---")
print(summary_table)
```
